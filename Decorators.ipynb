Topic: Decorators

Decorators are a powerful and flexible feature in python that allows you to modify the behavior of a function or a class method. 

They are commonly used to add functionality to functions or methods without modifying their actual code.

```python
### function copy
def welcome():
  return "Welcome to the advanced python course"

welcome()
```

```python
wel = welcome
print(wel)
```

```python
## closures
def main_welcome():
  msg = "Welcome"
  def sub_welcome_method():
    print("Welcome to the advanced python course")

    print("Please learn these concepts easily")

  return sub_welcome_method()
```
