Given a list of integers and an integer k. Write a Python function to rotate the list to the right by k positions without using slicing. A rotation shifts elements from the end of the list to the front.

Example:

Input: lst = [1, 2, 3, 4, 5], k = 2
Output: [4, 5, 1, 2, 3]

Input: lst = [10, 20, 30, 40, 50], k = 3
Output: [30, 40, 50, 10, 20]

```python
def rotate_list(lst, k):

  # Calculate the length of the list first
  n = len(lst)

  # Check if our list is empty or only have 1 element.  
  if n == 0:
    return lst 

  # Is k greater than length of a list?. Modulo operator gives remainder.
  k = k % n  

  # We define here helper function to to reverse list
  def reverse(start, end):
    while start < end:
      lst[start], lst[end] = lst[end], lst[start] # this is just swapping

      start += 1
      end -= 1

  # calling helper functions with arguments.
  reverse(0, n-1) 

  reverse(0, k-1) 

  reverse(k, n-1)

  return lst 

# outputs the solution
print(rotate_list([1, 2, 3, 4, 5], 2))
print(rotate_list([10, 20, 30, 40, 50], 3))
```
