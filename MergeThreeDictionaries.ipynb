```python
def merge_three_dictionaries(dict1, dict2, dict3):

  # Create a new dictionary that keeps copy of dict1
  merged_dict = dict1.copy()

  # Add a key-value pairs from dict2 and same for dict3
  for key, value in dict2.items():
    merged_dict[key] = value

  for key, value in dict3.items():
    merged_dict[key] = value

  # Finally return the new merged dictionary
  return merged_dict
  
print(merge_three_dictionaries({'a': 1, 'b': 2}, {'c': 3, 'd': 4}, {'e': 5, 'f': 6}))

print(merge_three_dictionaries({'a': 2, 'b': 2}, {'c': 5, 'd': 4}, {'e': 8, 'f': 9}))
```
