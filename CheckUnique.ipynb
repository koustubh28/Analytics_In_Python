```python
def check_unique(lst):
    # use set function to track elements
    new_list = set()
    
    for num in lst:
        if num in new_list:
            return False
            
        new_list.add(num)
        
    return True
```

```python
print(check_unique([1,2,3]))
```

```python
print(check_unique([1,2,3,3,4]))
```
