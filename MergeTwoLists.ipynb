Merge two Lists

```python
def merge_two_sorted_lists(list1, list2):
    
    merged_lists = []
    
    for i in list1:
        merged_lists.append(i)
        
    for j in list2:
        merged_lists.append(j)

    merged_lists.sort()  
    return merged_lists
            
print(merge_two_sorted_lists( [1, 3, 5], [2, 4, 6] ))
```
