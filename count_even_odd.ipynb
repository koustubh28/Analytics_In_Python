```python
def count_even_odd(lst):
    
    event_count = 0
    odd_count = 0
    
    for num in lst:
        if num % 2 == 0:
            event_count += 1 
            
        else:
            odd_count += 1 
            
    return event_count, odd_count
    
            
print(count_even_odd([1, 2, 3, 4, 5]))
print(count_even_odd([3, 2, 3, 8, 9, 57]))
```
