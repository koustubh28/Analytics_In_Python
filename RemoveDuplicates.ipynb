```python
def remove_duplicates(lst):

  # Initialize empty list
  new_list = []

  # Check if a number is not present in new list
  for number in lst:
    if number not in new_list:
      new_list.append(number)

  return(new_list)
```

```python
print(remove_duplicates([1,2,2,3]))
print(remove_duplicates([1,1,2,3,3]))
```
