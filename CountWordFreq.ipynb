Design a Python function named count_word_frequency to count the frequency of words in a sentence and store the counts in a dictionary.

```python
def count_word_freq(sentence):

  # Assign a new empty dictionary
  letDict = {}

  # Split the sentence to make it as an array
  words = sentence.split()
  
  # Loop through an array
  for word in words:
    # conditional logic to check if word present in empty dictionary
    if word in letDict:
      letDict[word] += 1

    else:
      # Let it remain 1
      letDict[word] = 1

  return letDict


print(count_word_freq("hello world hello"))

print(count_word_freq("I have forgotton that I have two breads."))
```
