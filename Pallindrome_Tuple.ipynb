Design a Python function named is_palindromic_tuple to check if a tuple is palindromic, meaning it reads the same forwards and backwards.

```python
def is_palindromic_tuple(tup):
    number = 0

    for digit in tup:
        number = number * 10 + digit

    original = number
    reversed_number = 0

    while number > 0:
        d = number % 10
        reversed_number = reversed_number * 10 + d
        number //= 10

    return original == reversed_number
```

```python
print(is_palindromic_tuple((1,2,3,2,1)))

print(is_palindromic_tuple((1,5,6,7)))

print(is_palindromic_tuple((11, 22, 33, 44, 55, 66, 77, 88, 99)))
```
