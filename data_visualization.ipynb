## SEABORN VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.info()
df.head()
```

```python
df['tip_percent'] = df['tip'] / df['total_bill']
df.describe()
```

```python
sns.boxplot(data=df, x='tip_percent')
```

```python
sns.boxplot(data=df, x='tip_percent', y='sex')
```

```python
sns.boxplot(data=df, y='day', x='tip_percent')
```

```python
days_ordered = ['Thur', 'Fri', 'Sat', 'Sun']
```

```python
sns.boxplot(data=df, y='day', x='tip_percent', order=days_ordered)
```

```python
df['day'] = pd.Categorical(df['day'], categories=days_ordered, ordered=True)
```

```python
df.info()
```

```python
sns.boxplot(data=df, y='day', x='tip_percent')
```

```python
sns.violinplot(data=df, y='day', x='tip_percent', hue='sex')
```

```python
sns.scatterplot(data=df, y='tip_percent', x='total_bill', hue='sex')
```

```python
sns.scatterplot(data=df, y='tip_percent', x='total_bill', hue='size')
```

```python
sns.scatterplot(data=df, y='tip_percent', x='total_bill', hue='size', style='sex')
```

```python
sns.lineplot(data=df, y='tip_percent', x='day')
```

```python
sns.lineplot(data=df, y='tip_percent', x='day', errorbar=None, estimator='median', color='red', label='MEDIAN')
sns.lineplot(data=df, y='tip_percent', x='day', errorbar=None, estimator='mean', color='blue', label='MEAN')
```

```python
sns.lineplot(data=df, y='tip_percent', x='day', errorbar=None, hue='sex')
sns.lineplot(data=df, y='tip_percent', x='day', errorbar=None, hue='sex', estimator='median', linestyle='dashed')
plt.title('Mean Solid; Median Dashed')
```

```python
sns.barplot(data=df)
```

```python
sns.barplot(data=df, y='tip', x='day', estimator='sum', errorbar=None)
```

```python
sns.barplot(data=df, y='day', x='tip', estimator='sum', errorbar=None)
```

```python
sns.barplot(data=df, y='day', x='tip', estimator='sum', errorbar=None, hue='sex')
```
