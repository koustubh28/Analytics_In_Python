## Working with groups in pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.info()
df.head()
```

## Exploring With Aggregations

```python
df.describe()
```

```python
df['tip'].mean()
```

```python
df['tip'].max()
```

```python
df[['total_bill', 'tip']].mean()
```

## Basic Filtering

```python
df[df['sex'] == 'Female']
```

```python
df[ df['sex'] == 'Male']
```

```python
df[ df['smoker'] == 'Yes']
```

```python
df[ df['size'] == 4]
```

```python
df[ df['sex'] == 'Female']['tip']
```

```python
print('mean', df[ df['sex'] == 'Female']['tip'].mean())
print('max', df[ df['sex'] == 'Female']['tip'].max())
```

```python
male_mask = df['sex'] == 'Male'
```

```python
print('mean', df[male_mask]['tip'].mean())
print('max', df[male_mask]['tip'].max())
```

## More Efficient: groupby

```python
df.groupby(by='sex')
```

```python
df.groupby('sex')['tip']
```
